### **Program Explanation:**

#### **1. `List` Example:**
   i) **`fruits`** is a list containing strings representing different fruits.
   
   ii) **`append`** method is used to add a new fruit **`("Cherry")`** to the list.

#### **2. `Tuple` Example:**
   i) **`colors`** is a tuple containing strings representing different colors.
   
   ii) Concatenation is used to create a new tuple **`(new_colors)`** by adding a new color **`("purple")`**.

#### **3. `Set` Example:**
   i) **`num`** is a set containing integers.
   
   ii) The add method is used to add a new number **`(6)`** to the set.

#### **4. `Dictionary` Example:**
   i) **`student_info`** is a dictionary representing information about a student.
   
   ii) A new key-value pair **`("course": "Pre - Engineering")`** is added to the dictionary

```python
def main():
    print("-----------------------------------------------------------------------------------")
    print("\n\t\t\t Python Data Structures Example:  ")
    print("\n-----------------------------------------------------------------------------------\n")
    # using List 
    fruits = ["apple", "banana", "grapes", "mango", "orange"]
    print("\nList of Fruits: \n", fruits)
    
    # using Tuple
    colors = ("Red", "Green", "Blue", "Yellow", "White")
    print("\nTuple of colors: \n", colors)
    
    # using set
    num = {1, 2, 3, 4, 5}
    print("\nSet of Numbers: \n", num)
    
    # Using Dictionary
    student_info = {"Name": "Ali", "Age": 24, "Grade": "A", "Class":"Inter", "City":"Karachi"}
    print("\nStudent Information: \n",student_info)
    
    print("\n-----------------------------------------------------------------------------------")

    # perform operation on list
    fruits.append("Cherry")
    print("\n\nAfter adding cherry to the list: ",fruits)
    
    # perform opterations on tuple
    new_colors = colors + ("purple",)
    print("\nAfter adding purple to the tuple: ", new_colors)
    
    # Perform operations on set 
    num.add(6)
    print("\nAfter adding 6 to the set: ", num)
    
    # Perform opteration on dictionary
    student_info["course"] = "Pre - Engineering"
    print("\nAfter adding course to the student dictionary: ", student_info)


if __name__ == "__main__":
    main()
```
